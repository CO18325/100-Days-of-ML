# Plotting Linear Graphs

```python
# if u pass directly the values of x and y then .plot() function will display the graph by default without specifying the .show() function
plt.plot([1,4,3],[4,9,1])
```

```python
#plt.plot(a,b,c,d)
##if you don't specify any coordinates in a multigraph, let say if we dont specify "c" in the graph --
## -- then the compiler will asume (0,1,2) on its own.
plt.plot([1,4,3],[4,9,1],[4,9,1])
```

```python
#multiple graphs in a single graph
plt.plot([1,4,3],[4,9,1],[0,3,2],[4,9,1])
```

```python
#we can plot graphs by specifying the coordinates in tuples,lists 
x = (1,4,8)
y = [4,6,7]

plt.plot(x,y)
```

```python
# Specifying the label's of the axis
plt.ylabel("label in y axis.")
plt.xlabel("label in x axis.")

#specifying the title in graph 
plt.title("TITLE")
plt.plot(x,y)
```

```python
#specifying more constraints to the line of graph
import matplotlib.style as stl

stl.use("ggplot")
#ggplot is style type in styles library

x = [1,4,7]
x2 = [3,5,8]
y = [4,1,8]
y2 = [6,0,2]
x3 = [3,6,9]
y3 = [5,7,0]

plt.plot(x,y,'g',label = 'Green-line',linewidth = 5 )
# -- 'g' & 'c' is specified for color of te line we want in our graph
plt.plot(x2,y2,'c',label = 'Green-line',linewidth = 5 )
#legend function to add legend to my graph
plt.legend()
#plt.grid function is used for enaabling grid structure in graphs
plt.grid(True,color = 'k')
```

```python
#displaying coordinates with dots,triangles and squares
#this third arguement specifies red color and dot notation of coordinates 
plt.plot(x,y, 'ro')
plt.plot(x2,y2, 'g^')
plt.plot(x3,y3, 'bs')
```

# Building trignometric graphs.

```python
import numpy as np
x = np.arange(0,4*np.pi,0.1) #start,stop,step positions of x graph
y = np.sin(x)
z= np.cos(x)
u= np.tan(x)
```

```python
# displaying sine graph
plt.figure(1)
plt.plot(y,'ro')
plt.ylim(-2,2)
# displaying cosine graph
plt.figure(2)
plt.plot(z,'gs')
plt.ylim(-2,2)
# displaying tan graph
plt.figure(3)
plt.plot(u,'b^')
plt.ylim(-2,2)
```

# Plotting a scatter plot.

```python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]
plt.scatter(x,y,label='skitscat',color='g',s=25,marker='^')
plt.title('Scatter_plot')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```

# Plotting a Stack plot.

```python
days = [1,2,3,4,5]

sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

plt.plot([],[],color='m',label="Sleeping",linewidth=5)
plt.plot([],[],color='c',label="Eating",linewidth=5)
plt.plot([],[],color='r',label="Working",linewidth=5)
plt.plot([],[],color='k',label="Playing",linewidth=5)

plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','r','k'])
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Stack_plot")
plt.legend()
```

# Plotting a Pie Chart

```python
plt.pie([7,2,2,13],
        labels=['sleeping','eating','working','playing'],
        colors=['c','m','r','b'],
        startangle=90,
        shadow=True,
        explode=(0,0,0.1,0),
       autopct='%1.1f%%')
plt.title("Pie_chart")
```
